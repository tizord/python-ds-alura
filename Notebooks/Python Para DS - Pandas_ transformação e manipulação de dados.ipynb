````python
import pandas as pd
import numpy as np
````

````python
dados = pd.read_json('/content/drive/MyDrive/Alura-Curso-07/dados_hospedagem.json')
dados.head()
````

````python
dados = pd.json_normalize(dados['info_moveis'])
dados
````

### 02. Dados numéricos

O método ```.explode()``` serve para transformar listas em linhas do df. Par aisso, é necessário passar o nome das colunas que precisam ser desagrupadas.

````python
colunas = list(dados.columns)
colunas
````

````python
dados = dados.explode(colunas[3:])
dados.reset_index(inplace=True, drop=True) # Não deixa coluna com índices antigos.
dados.head()
````

````python
dados.info()
````

````python
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
dados.info()
````

````python
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
dados[col_numericas] = dados[col_numericas].astype(np.int64)
dados.info()
````

````python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
dados.info()
````

````python
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$','').replace(',','').strip())
````

````python
dados['preco'] = dados['preco'].astype(np.float64)
dados.info()
````

````python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x: x.replace('$','').replace(',','').strip()).astype(np.float64)
dados.info()
````

### 03. Dados Textuais

Agora, iremos tokenizar os campos textuais. Para isso, iremos lidar com o método ```.str``` que permite ao pandas tratar o texto elemento a elemento. Esse resultado também poderia ser obtido usando o método ```.apply()```.

Vamos usar, também, o **regex**.

````python
dados['descricao_local'] = dados['descricao_local'].str.lower()
````

Retira todos os caracteres, menos de a-z minusculo e maiusculo, o hífen e a aspa simples.

````python
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']',' ', regex=True)
````

Retira todos os hífens que não fazem parte de um texto. Para isso, verifica se tem elemento antes e depois do hífen.

````python
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)',' ', regex=True)
````

````python
dados['descricao_local'] = dados['descricao_local'].str.split()
````

````python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"', '', regex=True)
````

````python
dados['comodidades'] = dados['comodidades'].str.split(',')
````

````python
dados['descricao_vizinhanca']
````

````python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.lower()
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']',' ', regex=True)
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('(?<!\w)-(?!\w)',' ', regex=True)
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split()
````

## Desafio 1

Analisar resultados de um evento com os clientes de uma empresa de vendas online.

#### Importar a base de dados que está no formato json e transformar em um dataframe.

````python
dados_vendas = pd.read_json('/content/drive/MyDrive/Alura-Curso-07/dados_vendas_clientes.json')
dados_vendas.head()
````

````python
dados_vendas = pd.json_normalize(dados_vendas['dados_vendas'])
````

````python
dados_vendas
````

### 02. Dados numéricos

#### Remover os dados em listas dentro do DataFrame;


````python
colunas_vendas = list(dados_vendas.columns)
colunas_vendas
````

````python
dados_vendas = dados_vendas.explode(colunas_vendas[1:])
dados_vendas.reset_index(inplace=True, drop=True)
dados_vendas
````

#### Verificar os tipos de dados;


````python
dados_vendas.info()
````

#### Identificar colunas numéricas e transformar a coluna numérica para o tipo numérico.;

````python
dados_vendas.head(3)
````

````python
dados_vendas['Valor da compra'] = dados_vendas['Valor da compra'].apply(lambda x: x.replace('R$ ','').replace(',','.').strip()).astype(np.float64)
````

````python
dados_vendas.info()
````

````python
dados_vendas.head(3)
````

### 03. Dados textuais

#### Manipular os textos presentes na coluna ```Cliente``` para que seja obtido como resultado os nomes dos clientes em letras minúsculas, com a ausência de caracteres especiais ou números.

````python
dados_vendas['Cliente'] = dados_vendas['Cliente'].str.lower()
````

````python
dados_vendas['Cliente'] = dados_vendas['Cliente'].str.replace('[^a-zA-Zà-úÀ=Ú0-9]',' ', regex=True)
````

````python
# dados_vendas['Cliente'] = dados_vendas['Cliente'].str.split()
````

````python
dados_vendas['Cliente']
````

````python
dados_vendas.head(3)
````

### 04. Dados de tempo

#### Transforme os dados de tempo para o o tipo datetime.

````python
dados_vendas['Data de venda'] = pd.to_datetime(dados_vendas['Data de venda'], format='%d/%m/%Y')
subset = dados_vendas.groupby(['Cliente'])['Valor da compra'].sum()
subset
````

## Desafio 2

Tratamento de dados com o objetivo de analisar o atraso no pagamento de aluguéis no condomínio de alguns moradores.

#### Importar a base de dados que está no formato json e transformar em um dataframe.

````python
dados_condominio = pd.read_json('/content/drive/MyDrive/Alura-Curso-07/dados_locacao_imoveis.json')
dados_condominio.head()
````

````python
dados_condominio = pd.json_normalize(dados_condominio['dados_locacao'])
dados_condominio
````

#### Remover os dados em listas dentro do DataFrame;


````python
colunas_aluguel = dados_condominio.columns
colunas_aluguel = list(colunas_aluguel)
colunas_aluguel
````

````python
dados_condominio = dados_condominio.explode(colunas_aluguel[1:])
dados_condominio.reset_index(inplace=True, drop=True)
dados_condominio
````

### 02. Dados numéricos

#### Verificar os tipos de dados;


````python
dados_condominio.info()
````

#### Identificar colunas numéricas e transformar a coluna numérica para o tipo numérico.

````python
dados_condominio['valor_aluguel'] = dados_condominio['valor_aluguel'].apply(lambda x: x.replace('$ ', '').replace(',','.').replace('reais', '').strip()).astype(np.float64)
````

````python
dados_condominio.info()
````

````python
dados_condominio.head(3)
````

### 03. Dados textuais

#### Manipular os textos na coluna apartamento para remover o texto ```(blocoAP) ``` do DataFrame.

````python
dados_condominio['apartamento'] = dados_condominio['apartamento'].str.replace(' \(blocoAP\)', '', regex=True)
````

````python
dados_condominio['apartamento']
````

### 04. Dados de data

 #### Transforme os dados das colunas ```datas_combinadas_pagamento``` e  ```datas_de_pagamento``` para o tipo ```datetime``` e busque uma forma de visualização de subconjunto que possa contribuir no objetivo de contar os dias de atraso do pagamento

````python
dados_condominio['datas_combinadas_pagamento'] = pd.to_datetime(dados_condominio['datas_combinadas_pagamento'], format='%d/%m/%Y')
dados_condominio['datas_de_pagamento'] = pd.to_datetime(dados_condominio['datas_de_pagamento'], format='%d/%m/%Y')
dados_condominio.head(5)
````

````python
dados_condominio['dias_atraso'] = dados_condominio['datas_de_pagamento'] - dados_condominio['datas_combinadas_pagamento']
media_atraso_pagamento = dados_condominio.groupby(['apartamento'])['dias_atraso'].mean()
````
