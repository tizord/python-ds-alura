**NaN**

O NaN siginifica Not a Number. Para podermos verificar se existes valores faltantes (ou nan), podemos usar a função

```python

```
Que irá trazer True/False se escontrar ou não um valor nan. Por fim, podemos somar esses valores, para sabermos quantos elementos nan existem.

Para tratar os valores nan, pode-se adotar diversas estratégias, onde uma delas é a interpolação, onde podemos pegar o valor imediatamente anterior e o valor imediatamente posterior e calcular a média entre eles. Por fim atribuimos esses valor ao valor nan.

## 04. Números Aleatórios

### Conceitos

A biblioteca numpy possui a função
```python
np.random.randint(min, max, qtd)
```
que gera números aleatórios inteiros.

Para gerar números racionais (float), podemos usar a função
```python
np.random.uniform(min, max, qtd)
```

**Reprodutibilidade**

A geração de números aleatórios não é 100% aleatória. Os números são pseudo-aleatórios e são calculados a partir de uma semente (**seed**), dessa forma, usando a mesma semente, deve-se chegar aos mesmos valores.
