### 04. Lidando com Outliers

Os outliers são valores atípicos, diferem do padrão do conjunto de dados significativamente. Para observar outliers, vamos usar o **boxplot**.

Agora, vamos entender 3 maneiras de identificar outliers:

1. **Z-score**: medida estatística que indica quantos múltiplos do desvio-padrão um dado está da média.

2. **3-sigmas**: medida estatística que considera que 99,7% dos dados estão dentro de 3 desvios-padrão da média.

3. **Análise de dispersão**: Técnica que utiliza gráficos, geralmente o boxplot.


````python
sns.boxplot(x=df_sem_nulos['cliente.tempo_servico'])
````

````python
Q1 = df_sem_nulos['cliente.tempo_servico'].quantile(0.25) # Selecionando o primeiro quartil
Q3 = df_sem_nulos['cliente.tempo_servico'].quantile(0.75) # Selecionando o terceito quartil
IQR = Q3 - Q1 # Calculando o intervalo interquartil
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
````

````python
outliers_index = (df_sem_nulos['cliente.tempo_servico'] < limite_inferior) | (df_sem_nulos['cliente.tempo_servico'] > limite_superior)
````

````python
df_sem_nulos[outliers_index]['cliente.tempo_servico']
````

````python
df_sem_out = df_sem_nulos.copy()
````

````python
df_sem_out[outliers_index]['cliente.tempo_servico']
````

````python
df_sem_out.loc[outliers_index, 'cliente.tempo_servico'] = np.ceil(
    df_sem_out.loc[outliers_index, 'conta.cobranca.Total'] /
    df_sem_out.loc[outliers_index, 'conta.cobranca.mensal']
)
````

````python
sns.boxplot(x=df_sem_out['cliente.tempo_servico'])
````

````python
df_sem_out[outliers_index][['cliente.tempo_servico', 'conta.cobranca.mensal', 'conta.cobranca.Total']]
````

Repare que tentamos corrigir os outliers. Porém, na coluna ```conta.cobranca.Total``` alguns valores já são outliers. Dessa forma, o jeito de corrigir é remover esses valores.

Primeiro vamos refazer o filtro de outliers.

````python
Q1 = df_sem_out['cliente.tempo_servico'].quantile(0.25) # Selecionando o primeiro quartil
Q3 = df_sem_out['cliente.tempo_servico'].quantile(0.75) # Selecionando o terceito quartil
IQR = Q3 - Q1 # Calculando o intervalo interquartil
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
````

````python
outliers_index_2 = (df_sem_out['cliente.tempo_servico'] < limite_inferior) | (df_sem_out['cliente.tempo_servico'] > limite_superior)
````

````python
df_sem_out[outliers_index_2]['cliente.tempo_servico']
````

````python
df_sem_out = df_sem_out[~outliers_index_2]
df_sem_out.reset_index(drop=True, inplace=True)
````

### 04. Lidando com Outliers

Foi informado que na base de dados dada a seguir há possíveis fraudes que podem ser identificadas através dos outliers no “valor da transação”.

Sua missão é construir uma função que recebe dois parâmetros, a coluna que você quer analisar se tem outlier e o DataFrame em questão.

O método que deve ser implementado para identificação do outlier é do intervalo interquartil.

````python
# criando um DataFrame com 30 transações aleatórias
df = pd.DataFrame({
    'ID da transação': range(1, 31),
    'Valor da transação': [100, 200, 150, 500, 300, 913, 250, 400, 200, 150,
                           200, 200, 400, 300, 150, 301, 805, 300, 400, 250,
                           150, 100, 500, 600, 200, 350, 100, 250, 800, 250],
    'Data da transação': pd.date_range(start='2022-01-01', end='2022-01-30', freq='D'),
    'Local da transação': ['São Paulo, Brasil', 'Rio de Janeiro, Brasil', 'Belo Horizonte, Brasil', 'São Paulo, Brasil',
                           'São Paulo, Brasil', 'Nova Iorque, EUA', 'São Paulo, Brasil', 'São Paulo, Brasil', 'São Paulo, Brasil',
                           'Rio de Janeiro, Brasil', 'São Paulo, Brasil', 'São Paulo, Brasil', 'São Paulo, Brasil', 'São Paulo, Brasil',
                           'São Paulo, Brasil', 'São Paulo, Brasil', 'Los Angeles, EUA', 'São Paulo, Brasil', 'São Paulo, Brasil', 'São Paulo, Brasil',
                           'São Paulo, Brasil', 'São Paulo, Brasil', 'São Paulo, Brasil', 'São Paulo, Brasil', 'São Paulo, Brasil', 'São Paulo, Brasil',
                           'São Paulo, Brasil', 'São Paulo, Brasil', 'Miami, EUA', 'São Paulo, Brasil']
})
````

````python
df
````

````python
def verifica_outliers(coluna: str, df: pd.DataFrame):
  Q1 = df[coluna].quantile(0.25) # Selecionando o primeiro quartil
  Q3 = df[coluna].quantile(0.75) # Selecionando o terceito quartil
  IQR = Q3 - Q1 # Calculando o intervalo interquartil
  limite_inferior = Q1 - 1.5 * IQR
  limite_superior = Q3 + 1.5 * IQR
  outliers_index = (df[coluna] < limite_inferior) | (df[coluna] > limite_superior)

  return df[outliers_index]
````

````python
verifica_outliers('Valor da transação', df)
````
