Pode-se remover colunas de um dataframe usando os métodos ```.drop()``` ou ```.pop()```. A diferença entre esses dois métodos são:

1. O método ```drop()``` necessita a flag ```inplace=True``` para que seja executado diretamento no dataframe original, já o método ```pop()``` é executado sobre o df original diretamente.
2. O método ```drop()``` permite remover diversas colunas de uma única vez ao passo que o método ```pop()```remove uma coluna por vez.
3. O método ```pop()``` retorna a coluna que está sendo removida e por isso pode-se armazenar essa coluna em alguma variável.

## 02. Agrupamento de dados

### Conceitos


O método ```.melt()```: Serve para transformar um df do formato wide (tabela pivô) para o formato long (normalizado). Em outras palavras, quando temos um df que tem diversas colunas com valores numéricos, por exemplo, várias colunas de anos, com seus valores e tantas outras com valores nominais, nós separamos as colunas nominais e as colunas de valores e usamos esse método .melt().

Como usar?

```python

dados.melt(
  id_vars=colunas_mantidas,
  value_vars=colunas_transformadas,
  var_name='nome_Col_nova',  
  value_name='Nome_col_valores'  
  )

```

Para fazer agragação, usando o método ```groupby()```devemos escolher como agrupar os dados. As possibilidades são:

```python
    mean() : calcula a média dos grupos
    sum() : calcula a soma dos valores do grupo
    size() : calcula tamanhos de grupo
    count() : calcula a contagem do grupo
    std() : desvio padrão de grupos
    var() : calcula a variação de grupos
    sem() : erro padrão da média dos grupos
    describe() : gera estatísticas descritivas
    first() : calcula o primeiro dos valores do grupo
    last() : calcula o último dos valores do grupo
    nth() : pega o n-ésimo valor do grupo
    min() : calcula o mínimo de valores do grupo
    max() : calcula o máximo de valores do grupo
  ```

## 03. Agrupamentos multi-index

### Conceitos

Para realizar agrupamentos multi-index, basta passar uma lista de colunas no método ```groupby()```.

```.xs()```: Cross-section. Serve para realizar uma seleção específica, em uma filtragem multi-index. Exemplo:

```python
gas_por_setor = dados.groupby(['Gas', 'Nível 1'])[['Emissão']].sum()

# Isso cria um df "dados_setor" que possui 2 níveis:
# 0 - Gas,
# 1 - Nível 1

gas_por_setor.xs('C02', level=0)
#Aqui, queremos identificar o gás CO2, que está no level 0 de índice

gas_por_setor.xs(('CO2', 'Mudança de Uso'), level=[0,1])

```

Poderíamos, também, determinar o maior valor, usando o método ```max()``` e o índice do maior valor, com o método ```idxmax()```

```python
gas_por_setor.xs('CO2', level = 0).max()
>>> 78321939
#Traz o maior valor, mas não diz o valor do Nível 1.

gas_por_setor.xs('CO2', level = 0).idxmax()
>>> 'Uso de terra'
# Agora traz o índice do maior valor no nível 1.

```

Quando tratamos de agrupamentos multi-index, temos que nos atentar para alguns detalhes. Podemos usar o método ``` .loc()  ```, porém, se passarmos apenas um valor, ele trará o primeiro nível do agrupamento multi-index. Para filtrar o valores de todos os níveis, devemos passar uma tupla com esses níveis.

Para que possamos selecionar valores específicos de algum dos níveis, devemos usar o método ``` .xs() ``` onde podemos selecionar o valor de um nível sem passar a tupla de todos os níveis.

Exemplo:

Imagina que 'NOx (t)' > 'Energia', 'Agropecuária'. Com o loc ou o xs ficaria:

```python
gas_por_setor.loc['NOx (t)']
>>> |'Agropecuária'|738219793|
>>> |'Energia'|532713|
>>> |'Processos Industriais'|3627139|

gas_por_setor.loc[('NOx (t)', 'Energia')]
>>> 532713 #Valor associado a esse conjunto

gas_por_setor.xs('Energia', level = 1)
>>> tabela com todos os gases com valor de 'Energia' no nível 1
```

Podemos fazer agrupamento em tabelas já agrupadas, da seguinte forma

```python

gas_por_setor.groupby(level = 0).idxmax()
```

Esse código, agrupa pelo nível 0 e traz o idxmax, ou seja, o nome da atividade (nivel 1) com o maior valor de emissão.

Para trazer o valor, em vez do nome, trocamos o ```idxmax()``` por ```max()```


O método ***insert()***: Capaz de inserir uma nova coluna em uma tabela.

tabela.insert(


````python
gas_por_setor = emissoes_por_ano.groupby(['Gás', 'Nível 1 - Setor'])[['Emissão']].sum()
gas_por_setor.groupby(level = 0).idxmax() # Agrupamos pelo nível 0 e trazemos o índice do maior valor.
gas_por_setor.groupby(level = 0).max() # Agrupamento pelo nível 0 e trazemos o maior valor
valores_max = gas_por_setor.groupby(level = 0).max().values
tabela_sumarizada = gas_por_setor.groupby(level = 0).idxmax()
tabela_sumarizada.insert(1, 'Quantidade de emissão', valores_max)
gas_por_setor.swaplevel(0, 1).groupby(level=0).idxmax()
````

## 04. Unindo dados

````python
populacao_estados = pd.read_excel('/content/drive/MyDrive/Alura-Curso-06/POP2022_Municipios.xls', header=1, skipfooter=34)
populacao_estados.groupby('UF').sum(numeric_only=True)
````

````python
populacao_estados['POPULAÇÃO'].astype(int)
````

````python
populacao_estados[populacao_estados['POPULAÇÃO'].str.contains('\(', na=False)]
````

````python
populacao_estados = populacao_estados.assign(populacao_sem_parenteses = populacao_estados['POPULAÇÃO'].replace('\(\d{1,2}\)','', regex=True), populacao = lambda x: x.loc[:, 'populacao_sem_parenteses'].replace('\.', '', regex=True))
populacao_estados[populacao_estados['POPULAÇÃO'].str.contains('\(', na=False)]
````

````python
populacao_estados = populacao_estados.astype({'populacao': 'int64'})
populacao_estados.info()
````

````python
populacao_estados = populacao_estados.groupby('UF')[['populacao']].sum().reset_index()
populacao_estados
````

````python
emissao_estados = emissoes_por_ano[emissoes_por_ano['Ano'] == 2021].groupby('Estado')[['Emissão']].sum().reset_index()
dados_agrupados = pd.merge(emissao_estados, populacao_estados, left_on='Estado', right_on='UF')
dados_agrupados.head()
````

````python
dados_agrupados.plot(x='populacao', y='Emissão', kind='scatter', figsize=(8,6));
````

````python
import plotly.express as px
````

````python
px.scatter(data_frame=dados_agrupados, x='populacao', y='Emissão', text='Estado', opacity=0)
````

````python
dados_agrupados = dados_agrupados.assign(emissao_per_capita = dados_agrupados['Emissão']/dados_agrupados['populacao']).sort_values('emissao_per_capita', ascending=False)
dados_agrupados
````

````python
px.bar(data_frame=dados_agrupados, x='Estado', y='emissao_per_capita')
````

````python
px.scatter(data_frame=dados_agrupados, x='populacao', y='Emissão', text='Estado', size='emissao_per_capita')
````

````python
dados_agrupados['Emissão'] = dados_agrupados['Emissão'].apply(lambda x: x if x>=0 else -x)
dados_agrupados = dados_agrupados.assign(emissao_per_capita = dados_agrupados['Emissão']/dados_agrupados['populacao']).sort_values('emissao_per_capita', ascending=False)
dados_agrupados
````

````python
px.scatter(data_frame=dados_agrupados, x='populacao', y='Emissão', text='Estado', size='emissao_per_capita')
````

o método .assign permite criar colunas a partir de colunas criadas dentro da própria função assign.

O Pandas possui 3 métodos para unir dataframes:
* ```concat()```:
```python
pd.concat([dataframe1, dataframe2, ...], axis=0/1)
```
* ```join()```:
```python
dataframe1.join(dataframe2, on='nome_coluna', how='left')
```
* ```merge()```:
```python
dataframe1.merge(dataframe2, on=None, how='inner', left_on='nome_coluna_dataframe1', right_on='nome_coluna_dataframe2')
```
